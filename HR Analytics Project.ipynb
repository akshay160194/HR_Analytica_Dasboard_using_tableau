# Data Pre-Processing

```python
# Performing EDD for numerical variables
df.describe()
```

```python
#looking for null values
df.isnull().sum()
```

```python
# dropping duplicate
df = df.drop_duplicates()
#dropping NaN values
df = df.dropna()
```

```python
df.shape
```

```python
# Saving the transformed data
df.to_csv('HR_dataset.csv')
```

## Plotting Correlational Matrix

```python
# for those who prefer visualization plotting a correlation heatmap
plt.figure(figsize = (20,14))
plt.title('HR Correlational Matrix')
corr = df.corr()
sns.heatmap(corr,xticklabels=corr.columns,yticklabels=corr.columns,linewidths=.1,cmap="coolwarm", annot = True)
plt.show()
```

No higly correlated predicter variable present in the dataset

## Looking at Categorical Variable

```python
# Overtime
sns.countplot(data=df, x='OverTime')
plt.title('OverTime')
plt.show()
```

```python
# MaritalStatus 
sns.countplot(data=df, x='MaritalStatus')
plt.title('MaritalStatus')
plt.show()
```

```python
# JobRole 
plt.figure(figsize=(10,5))
sns.countplot(data=df, y='JobRole')
plt.title('JobRole')
plt.show()
```

```python
# Gender
sns.countplot(data=df, x='Gender')
plt.title('Gender')
plt.show()
```

```python
# EducationField 
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='EducationField')
plt.title('EducationField')
plt.show()
```

```python
# Department
sns.countplot(data=df, x='Department')
plt.title('Department')
plt.show()
```

```python
# BusinessTravel
sns.countplot(data=df, x='BusinessTravel')
plt.title('BusinessTravel')
plt.show()
```

```python
# OverTime Vs Age
sns.boxplot(x="OverTime",y="Age",data=df)
plt.title('OverTime Vs Age')
plt.show()
```

## Salary Analysis

```python
# Monthly Income Distribution
sns.histplot(data=df, x='MonthlyIncome', kde=True)
plt.title('Monthly Income Distribution')
plt.show()
```

```python
# Monthly Income Vs Job Role
plt.figure(figsize=(10,5))
plt.xticks(rotation = 90)
sns.boxplot(data=df, x='JobRole', y='MonthlyIncome')
plt.title('Monthly Income Vs Job Role')
plt.show()
```
